# Boring- Channel- AF3

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/rufus/OneDrive/Desktop/GAMEEMO/(S01)/Preprocessed EEG Data/.csv format/S01G1AllChannels.csv")
df.head()
```

```python
del df['Unnamed: 14']
```

```python
df
```

```python
import numpy as np
from ssqueezepy import cwt
from ssqueezepy.visuals import plot, imshow

#%%# Make `x` & plot #########################################################
t = np.linspace(0, 1, 2048, 1)
x = np.array(df['AF3'])

plt.axis('off')
plot(x, title="x(t) | t=[0, ..., 1], %s samples" % len(x), show=1)

plt.axis('off')
#%%# Take CWT & plot #########################################################
Wx, scales = cwt(x, 'morlet')
imshow(Wx, yticks=scales, abs=1, title="abs(CWT) | Morlet wavelet", ylabel="scales", xlabel="samples")
```
